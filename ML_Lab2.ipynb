```python
import sklearn.model_selection as model_selection

data = pd.read_csv("salary_data.csv")

x = data['YearsExperience']
y = data['Salary']

n = 0.1
for i in range(9):
  x_train, x_test, y_train, y_test = model_selection.train_test_split(x, y, test_size=1-n)

  print("X =", round(n*100))
  print("Y =", round((1-n)*100))

  print ("TRAIN_x:")
  print(x_train)
  print ("\nTRAIN_y:")
  print(y_train)

  print("\nTEST_x:")
  print(x_test)
  print ("\nTEST_y:")
  print(y_test)
 
  n = n + 0.1
```

```python
#overall
x = data['YearsExperience']
d = data['Salary']
overall = x + d
print("Mean:",overall.mean())
print("Median:",overall.median())
print("Mode:",overall.mode())
print("Standard Deviation:",overall.std())
```

# **Part 2**

```python
import matplotlib.image as img
imageMatrix = img.imread('image.jpg')#reading image and storing in the matrix (3d)

#converting from the 3D matrix to 2D matrix
imageMatrix = imageMatrix.reshape(imageMatrix.shape[0],imageMatrix.shape[1]*imageMatrix.shape[2])
print(imageMatrix)

df = pd.DataFrame(imageMatrix)#converting the matrix to data frame
 
df.to_csv('file1.csv',index=False)#saving the dataframe as csv file
```

```python
len(df)#length of dataframe gives the number of rows 
new_df = pd.read_csv('file1.csv',nrows=len(df)-1,usecols=range(len(df.columns)-1))
print(new_df)
```

```python
def birth_age(age):
  today=date.today()
  age=today.year - age.year - ((today.month, today.day) <(age.month, age.day)) 
  return age
print("enter you birth date:")         
year=int(input("year: "))
month=int(input("month: "))
day=int(input("day: "))
age=birth_age((date(year,month,day)))
print(age,"years")
```
